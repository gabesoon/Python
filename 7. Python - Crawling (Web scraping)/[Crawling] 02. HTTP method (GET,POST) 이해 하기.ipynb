# HTML (Hyper Text Markup Language)

Hyper Text Transfer Protocol : HTML 문서 등의 리소스를 전송하는 프로토콜

* Protocol : 미리 정의된 규약
* Hyper Text : HTML문서 : 우리가 보는 모든 웹페이지

HTTP를 요청하는 방식은 여러가가 있다. 
python을 이용한 Crawling에 가장 많이 사용되는 요청은 2가지이다.

# Get 요청

- 가장 많이 사용되는 방식
- 주로 리소스 요청에 사용된다.

- 데이터를 URL에 포함하여 전달하는 방식

# Post 요청

- 데이터를 Form data에 포함하여 전달하는 방식
- 주로 로그인 할때 많이 사용된다.(99%)

특정 웹페이지에 로그인 할때, 입력한 ID와 PW가 Form Data 라는 형식으로 전송했던 것을 확인할 수 있다.

---

- 즉, 일반적으로 크롤링을 하기 위해서는 주로 get 방식을 통해 원하는 데이터를 가져오게 될 것이다.
- 간혹가다 데이터를 가져오기 위해 Post를 구성해야 하는 경우가 발생할 수 있다.
- 혹은 로그인을 한 뒤에 데이터를 크롤링 해야만 하는 경우가 있다. 이때는 Post 방식을 통하게 될 것이다.
