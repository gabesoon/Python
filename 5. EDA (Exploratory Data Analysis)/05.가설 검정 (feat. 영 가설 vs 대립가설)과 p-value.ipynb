# 통계적 가설 검정
---
- 가설 검정 : 특정 가설에 대하여 데이터(샘플)를 바탕으로 통계적으로 유의미한 가설인지 아닌지를 확인하는 과정<br><br>
    - 엄밀히 말해 세상에 존재하는 모든 데이터는 (특별한 몇몇 경우 제외) 모두 샘플이다. 즉, 모집단을 정확히 알 수 없다.<br><br>
    - e.g) 전세계 모든 남성의 키의 평균은 얼마일까?<br><br>
    - 전세계 모든 남성을 모아서 키를 측정하는 것은 불가능에 가까움<br><br>
    - 이를 대략적으로 파악하기 위해, 샘플 데이터를기반으로 모집단(전세계 남성)의 평균키를 가늠하는 것<br><br>

- 통계적으로 유의하다는 것의 의미?<br><br>
    - 통계적 검증(실험)의 결과가 확률적으로 단순한 우연이 아니라고 인정될 정도의 의미가 있다는 것.<br><br>
    - 즉, 검증 결과가 우연이라고 하기에는 너무 확률이 작다는 뜻 (p-value로 판단)<br><br>

# 영 가설 & 대립 가설
---
- 영 가설 (null hypothesis)<br><br>
    - 특별한 증거가 없으면 *참으로 추정되는* 가설<br><br>
    - 검증하고자 하는(관심이 있는) 대상이 아닌 가설<br><br> 
    - 즉, 통계적 검정을 통해 영 가설을 기각하고자 하는 것이 실험의 목적<br><br>
    - **$H_{0}$** 로 표기한다.<br><br>
    - 모집단에 대한 특성을 등호 (같다 or 같지 않다)로만 표현<br><br>
    
- 대립 가설 (alternative hypothesis)
    - 특별한 증거(통계적 유의함)이 없으면 거짓으로 추정되는 가설<br><br>
    - 검증하고자 하는 (관심이 있는) 대상이 되는 가설<br><br>
    - **$H_{1}$ 혹은 $H_{a}$**로 표기한다.<br><br>
    - 모집단에 대한 특성을 부등호로 표기한다. (단측, 양측 검정 모두 가능)<br><br>
---

<영 가설과 대립가설 예시><br><br>

- 기본적으로 검증이 완료되어 판단이 내려지기 전까지는 영 가설이 참이라는 전체하에 판단한다.<br><br>

영 가설($H_{0}$) : 대한민국 성인 남성 키의 평균은 173cm이다.<br><br>


대립 가설($H_{1}$) : 대한민국 성인 남성 키의 평균은 173cm과 같지 않다. (양측 검정)<br><br>
- 양측 검정 : 기준대비 미만 혹은 초과를 모두 검정<br><br>

- 만약 검증 결과 성인 남성의 키 평균이 173cm 보다<br><br>
    - 크다면 영 가설을 기각한다.<br><br>
    - 작다면 영 가설을 기각한다.<br><br>
    - 이 경우 대립가설이 참이 된다.<br><br>



- 만약 검증 결과 성인 남성의 키 평균이 173cm라면 대립가설을 기각한다.<br><br>
    - 이 경우 대립가설이 거짓이 된다.<br><br>

# p-value (유의확률)
---

- 영 가설이 맞다고 가정했을때, 실제 실험에서 다른 결과가 관측될 확률<br><br>
    - 즉, 실험 결과가 영 가설이 참일때 나올 수 있는 데이터가 아닌 것에 대한 확률<br><br>
- p-value가 작을 수록 영 가설을 기각할 근거가 강력해진다.<br><br>
    - 보통 0.05 혹은 0.01 미만이면 영 가설을 기각 한다.<br><br>
    - 즉, 실험 결과가 영 가설이 참일때의 값과 다를 확률이 95% 혹은 99%인 것.<br><br>
---

e.g)<br><br>
영 가설($H_{0}$) : 대한민국 성인 남성 키의 평균은 173cm이다.<br><br>

- 그런데, 실제로 수집한 데이터(샘플)로 가설 검정을 한 결과 남성 키의 평균이 180cm이 나왔다.<br><br>
- 이 경우, 영 가설을 참(남성 키 평균이 173)이라고 했으므로 실제 결과와 다르기 때문에 영 가설이 거짓이라 판단한다.(영 가설 기각)<br><br>


---

![image](https://user-images.githubusercontent.com/74717033/134213551-fe373833-b578-44dd-892b-ac28088cdff8.png)

- 위의 그림처럼 **'영 가설이 맞다고 가정할 때 기대할 수 있는 분포'** 에 걸쳐 있거나(빗금 영역) 혹은 그것을 초과하는 범위에 데이터가 분포하고 있다면 **"영 가설이 참이 아니다."** 라고 판단할 것이다.<br><br>

- 영 가설이 기각되었을때, 대립가설이 참이다. 라고 무조건 판단하면 안된다. (특히, 단츠 검정) <br><br>



- e.g) 
- 영 가설($H_{0}$) : 성인 남성의 키는 성인 여성의 키와 같다.<br><br>

- 대립 가설($H_{1}$) : 성인 남성의 키는 성인 여성의 키보다 크다. (단측 검정)<br><br>


- 검증 결과가 영 가설을 기각했을때는 "성인 남성의 키가 성인 여성의 키가 같지 않다" 가 참인 명제이다.<br><br>


- 즉, 성인 남성의 키가 성인 여성의 키보다 클 수도, 작을 수도 있기 때문에 위의 대립가설이 무조건적으로 참이 되지는 않는다.<br><br>

## p-value 예시
---

- 영 가설($H_{0}$) : 성인 남성의 키는 160cm일 것이다.<br><br>

- 대립 가설($H_{1}$) : 성인 남성의 키는 160cm 이상일 것이다.<br><br>

- 검증 결과(성인 100명 샘플) 성인 남성의 키 평균은 175cm, 표준편차는 1cm 였다.<br><br>

![image](https://user-images.githubusercontent.com/74717033/134215587-000e29f4-decd-47a5-bebc-93f26b12c925.png)

- 영 가설이 맞다면, 위의 그림과 같이 160인 데이터 중심에서 데이터들이 관측되었어야 한다.<br><br>

- 하지만 실제 결과는 우측의 175 부근에서 관측되었다.<br><br>
- 실질적으로 이는 불가능한 수치는 아니지만, 남자키 평균 160cm가 참이라면 수집한 데이터도 그 평균이 근사치가 나와야 할 것이다.<br><br>
- 하지만 그러하지 않으므로 영가설을 기각 한다.

## p-value 예시2
---

- 두 집단의 데이터를 비교하는 가설일 경우

- 영 가설($H_{0}$) : 성인 남성의 키는 성인 여성의 키와 같을 것이다.<br><br>

- 대립 가설($H_{1}$) : 성인 남성의 키는 성인 여성의 키보다 작다.<br><br>

---
![image](https://user-images.githubusercontent.com/74717033/134215871-b73fc63c-b188-4726-93da-8dd3390fd9da.png)

- 영 가설이 참이라면, 여성의 키분포와 남성의 키분포가 겹치는 부분이 많을 것이다. (혹은 완전 일치)<br><br>
- 하지만 위 그림과 같이 겹치는 부분이 적으면 적을수록 두 집단간의 차이가 있음을 알 수 있다.
---

- 실제 가설 검정에서는 위쪽 그래프 처럼 두 집단의 분포를 각각 그리기 보다는 두 집단의 데이터를 하나로 만들어 아래와 같은 그래프로 그린다.<br><br>
    - 일반적으로 남성의 키가 여성의 키보다 크다는 전제하에 (남성키 - 여성키) 한 데이터의 분포를 그려보는 것.<br><br>
    - 영 가설이 참이라면 해당 데이터의 분포는 모두 0 부근에 몰려있을 것이다.<br><br>
    - 영 가설이 거짓이라면 해당 데이터의 분포는 0부근이 아닌 다른 부분에 위치할 것이다. (p-value 기준으로 본다면 양 끝단 혹은 그 이상의 부분)
