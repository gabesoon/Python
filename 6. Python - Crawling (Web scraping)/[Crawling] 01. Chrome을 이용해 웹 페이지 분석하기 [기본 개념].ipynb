# 개발자 도구

- 파이썬을 통한 웹 크롤링에 기본이 되는 개념
- 특정 브라우져 (Chrome이 보편적이므로 Chrome 기준으로 정리) 가 html 정보를 받아와서 어떤식으로 사용자가 보기 좋게 렌더링하는지 확인할 수 있다.
- 파이썬 웹 크롤링을 위해 개발자 도구에서 살펴봐야 할 탭은 2개 이다. 
    - Elements : 크롤링할 데이터가 어디에 어떤 형식으로 위치하는지 확인 가능
    - Network : 웹페이지를 불러올때 어떤 요청을 받아서 어떤 형식으로 처리하였는지 확인 가능

## 개발자 도구 열기

- 크롬창 오픈 -> '보기(view)' -> '개발자(Developer)' -> '개발자 메뉴(Developer Tools)'(리눅스)
- 크롬창 오픈 -> '우측 점 3개 클릭' -> '도구 더보기' -> '개발자 도구' (단축키 Ctrl+Shift+I)  (윈도우)

## 특정 요소의 elements 확인하기

- 개발자 도구에서 1번의 모양 클릭 or Ctrl+Shift+C 를 눌러 element를 search 할 수 있다.
- 확인하고 싶은 위치를 클릭하면, 개발자 도구 창의 Elements에 해당 element의 정보가 어떤 것인지 음영으로 표시된다.
- 즉, 원하는 데이터를 클릭했을때 해당 데이터가 html에서 어떤 테그와 어떤 속성을 갖는지를 elements 탭에서 확인할 수 있다.
