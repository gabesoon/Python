# 태그, 속성, 값

- HTML 문서의 기본 블락
- 브러우져에게 어떻게 렌더링(화면에 표시) 될지 전달 하는 기능

```python
<태그명 속성1="속성값1" 속성2="속성값2">Value</태그명>
```
과 같이 <태그명> : 시작점 , </태그명> : 종료지점 을 의미한다.
태그명 사이에 들어갈 값이 있으면 Value 부분에 입력된다.

태그명 사이에 들어갈 값이 없으면 아래와 같이 표현되기도 한다.
```python
<태그명 속성1="속성값1" 속성2="속성값2"/>
```

e.g)
```python
<h3 class="tit_view" data-translation="true">공공기관, 민간 데이터 구매 늘린다</h3>
```



출처 : https://news.v.daum.net/v/20210706153019736

- p(paragraph) : p 테그 안의 값을 한 문단으로 표시하도록 하는 기능
- div : 논리적으로 여러 태그들을 묶어서 그룹핑 하는 기능

등 다양한 종류의 테그가 있다. 

단순히 테그가 어떤 것인지 인지 하는 수준이면 크롤링을 하는데에 충분하다.

**크롤링을 통해 추출하고자 하는 데이터의 대부분은 값 에 있다.**

즉, 원하는 값이 있는 태그만 찾을 수 있으면, 쉽게 데이터를 추출할 수 있다.
